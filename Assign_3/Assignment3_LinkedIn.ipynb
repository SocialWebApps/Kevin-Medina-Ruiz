## Step #6: Simple normalization of company suffixes from address book data

In this step I go through my connection list and normalize their respective employers to then get a frequency count of my connection's employers.

```python
import os # provides current path - I substituted with absolute path
import csv
from collections import Counter
from operator import itemgetter
from prettytable import PrettyTable

# XXX: Place your "Outlook CSV" formatted file of connections from 
# http://www.linkedin.com/people/export-settings at the following
# location: resources/ch03-linkedin/my_connections.csv

#CSV_FILE = os.path.join("resources", "ch03-linkedin", 'my_connections.csv')
CSV_FILE = "C:\\Users\\Kevin\\Desktop\\Social Web\\LinkedIn\\my_connections2.csv"
# Define a set of transforms that converts the first item
# to the second item. Here, we're simply handling some
# commonly known abbreviations, stripping off common suffixes, 
# etc.

transforms = [(', Inc.', ''), (', Inc', ''), (', LLC', ''), (', LLP', ''),
               (' LLC', ''), (' Inc.', ''), (' Inc', '')]

csvReader = csv.DictReader(open(CSV_FILE), delimiter=',', quotechar='"')
contacts = [row for row in csvReader]
companies = [c['Company'].strip() for c in contacts if c['Company'].strip() != '']

for i, _ in enumerate(companies):
    for transform in transforms:
        companies[i] = companies[i].replace(*transform)

pt = PrettyTable(field_names=['Company', 'Freq'])
pt.align = 'l'
c = Counter(companies)
[pt.add_row([company, freq]) 
 for (company, freq) in sorted(c.items(), key=itemgetter(1), reverse=True) 
     if freq > 0] 
print pt 
```

## Step #8: Normalizing Job Titles - (Option 2)

In this step I normalized the table created in step #7. The normalization rules I used are found are key-value pairs inside the "normalization" dictionary below.

```python
import os
import csv
from operator import itemgetter
from collections import Counter
from prettytable import PrettyTable

# XXX: Place your "Outlook CSV" formatted file of connections from 
# http://www.linkedin.com/people/export-settings at the following
# location: resources/ch03-linkedin/my_connections.csv

#CSV_FILE = os.path.join("resources", "ch03-linkedin", 'my_connections.csv')
CSV_FILE = "C:\\Users\\Kevin\\Desktop\\Social Web\\LinkedIn\\my_connections2.csv"

#NORMALIZATION RULES
normalization = {
                 'Analyst': 'Analyst',
                 'Accountant': 'Accountant',
                 'Coach': 'Coach',
                 'Consultant' :'Consultant',
                 'Developer': 'Developer',
                 'HR': 'Human Resources',
                 'Human': 'Human Resources',
                 'Vice': 'Vice President',
                 'VP': 'Vice President',
                 'V.P.': 'Vice President',
                 'Engineer': 'Engineer',
                 'Engineering': 'Engineer',
                 'Engineer,': 'Engineer',
                 'Estimating': 'Estimating',
                 'Intern': 'Intern',
                 'Internship': 'Intern',
                 '(Intern)': 'Intern',
                 'Designer': 'Designer',
                 'Receptionist': 'Receptionist',
                 'Desk': 'Receptionist',
                 'Tutor': 'Tutor',
                 'Technician': 'Technician',
                 'Executive': 'Executive',
                 'Sales': 'Sales',
                 'Sale': 'Sales',
                 'Recruiter': 'Recruiter',
                 'Recruitment': 'Recruiter',
                 'Recruiting': 'Recruiter',
                 'Professor': 'Professor',
                 'Professor,': 'Professor',
                 'Nurse': 'Nurse',
                 'Manager': 'Manager',
                 'MANAGER': 'Manager',
                 'Owner': 'Owner',
                 'owner': 'Owner',
                 'Co-Owner': 'Owner',
                 'Student': 'Student',
                 'Officer': 'Officer'
                }

#Open csv file
csvReader = csv.DictReader(open(CSV_FILE), delimiter=',', quotechar='"')
contacts = [row for row in csvReader]

# Read in a list of job titles
titles = []
for contact in contacts:
    titles.extend([t.strip() for t in contact['Job Title'].split('/')
                  if contact['Job Title'].strip() != ''])

# Nomralize job titles with our dictionary 
for i, _ in enumerate(titles):
    split_title = titles[i].split()
    for y in range(len(split_title)):
        if(normalization.has_key(split_title[y])):
              titles[i] = normalization[split_title[y]]


# Print out a table of titles sorted by frequency
pt = PrettyTable(field_names=['Title', 'Freq'])
pt.align = 'l'
c = Counter(titles)
[pt.add_row([title, freq]) 
 for (title, freq) in sorted(c.items(), key=itemgetter(1), reverse=True) 
     if freq > 1]
print pt
```
