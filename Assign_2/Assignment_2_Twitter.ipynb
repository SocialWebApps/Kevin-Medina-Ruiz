### Step #1: Authorizing app to access Twitter API

This is where you must enter your OAuth credentials. The expected output is a twitter object as shown in the expected output cell.

```python
import twitter

CONSUMER_KEY = ''  
CONSUMER_SECRET = ''
OAUTH_TOKEN = ''
OAUTH_TOKEN_SECRET = ''

auth = twitter.oauth.OAuth(OAUTH_TOKEN, OAUTH_TOKEN_SECRET,CONSUMER_KEY, CONSUMER_SECRET)

twitter_api = twitter.Twitter(auth=auth)

print twitter_api
print
```

#### Expected Output:

The output should just describe a twitter object like the following: < twitter.api.Twitter object at 0x0000000003D5A748 >

### Step #2: Retrieve trending topics from 5 different locations using the WOE ID system

In this step we will grab the trending topics in 5 different locations. I used the WOE ID assigned by Yahoo. 
For this assignment I picked the following locations: 

1. LOS ANGELES, CALIFORNIA 
2. USA
3. UNITED KINGDOM
4. CANADA
5. WORLDWIDE

```python
# See https://dev.twitter.com/docs/api/1.1/get/trends/place and
# http://developer.yahoo.com/geo/geoplanet/

LA_WOE_ID = 2442047
US_WOE_ID = 23424977
UK_WOE_ID = 23424975
CANADA_WOE_ID = 23424775
WORLD_WOE_ID = 1

# Prefix ID with the underscore for query string parameterization.
# Without the underscore, the twitter package appends the ID value
# to the URL itself as a special case keyword argument.

la_trends = twitter_api.trends.place(_id=LA_WOE_ID)
us_trends = twitter_api.trends.place(_id=US_WOE_ID)
world_trends = twitter_api.trends.place(_id=WORLD_WOE_ID)
uk_trends = twitter_api.trends.place(_id=UK_WOE_ID)
canada_trends = twitter_api.trends.place(_id=CANADA_WOE_ID)


print la_trends
print
print us_trends
print
print world_trends
print
print uk_trends
print
print canada_trends
```

#### Expected Output:

### Step #3: Pretty printing the results using JSON

In this step we will organize the output into a more readable format using the JSON package.

```python
import json

print json.dumps(la_trends, indent=1)
print
print json.dumps(us_trends, indent=1)
print
print json.dumps(world_trends, indent=1)
print
print json.dumps(uk_trends, indent=1)
print
print json.dumps(canada_trends, indent=1)
print
```

#### Expected Output:

### Step #4: Finding the intersection of 5 trending topic sets

In this step we will find the common trends in all 5 locations. This is done using sets and their built in 'intersection' function. First we create a set from the data retrieved in steps 2 and 3. A set will give us only unique elements. Then we will use the built in intersection function to retrieve the common trends in all 5 locations.

```python
la_trends_set = set([trend['name'] 
                        for trend in la_trends[0]['trends']])

us_trends_set = set([trend['name'] 
                        for trend in us_trends[0]['trends']])

world_trends_set = set([trend['name'] 
                        for trend in world_trends[0]['trends']])

uk_trends_set = set([trend['name'] 
                        for trend in uk_trends[0]['trends']])

canada_trends_set = set([trend['name'] 
                     for trend in canada_trends[0]['trends']]) 

common_trends = la_trends_set.intersection(us_trends_set,canada_trends_set,world_trends_set,uk_trends_set)

print common_trends
```

### Step #4: Finding the locations of the common trending topics in the 5 locations.

In step #4 we found the trending topics in all 5 locations. ***NOTE: There not be any common trends, depending on what is happening at those locations when this notebook is being used.***

For me there were only two common trends, they were '#HappyMatturday' and '#MakeSomeonesDayIn3Words'.
We will work with these two trends for the rest fo the assignment.

Now we will search for these specific hashtags to retrieve the tweets associated with them. I limited my search to 5 objects of 100 results each.

```python
# Import unquote to prevent url encoding errors in next_results
from urllib import unquote

# XXX: Set this variable to a trending topic, 
# or anything else for that matter. The example query below
# was a trending topic when this content was being developed
# and is used throughout the remainder of this chapter.

q = '#HappyMatturday' 
q2= '#MakeSomeonesDayIn3Words'
count = 500

search_results = twitter_api.search.tweets(q=q, count=count)
search_results2 = twitter_api.search.tweets(q=q2, count=count)

statuses = search_results['statuses']
statuses2 = search_results2['statuses']

# Iterate through 5 more batches of results by following the cursor

for _ in range(5):
    print "Length of statuses", len(statuses)
    try:
        next_results = search_results['search_metadata']['next_results']
    except KeyError, e: # No more results when next_results doesn't exist
        break
        
    # Create a dictionary from next_results, which has the following form:
    # ?max_id=313519052523986943&q=NCAA&include_entities=1
    kwargs = dict([kv.split('=') for kv in unquote(next_results[1:]).split("&") ])    

    
    search_results = twitter_api.search.tweets(**kwargs)
    statuses += search_results['statuses']
    
# Iterate through 5 more batches of results by following the cursor

for _ in range(5):
    print "Length of statuses2", len(statuses2)
    try:
        next_results2 = search_results2['search_metadata']['next_results']
    except KeyError, e: # No more results when next_results doesn't exist
        break
        
    # Create a dictionary from next_results, which has the following form:
    # ?max_id=313519052523986943&q=NCAA&include_entities=1
    kwargs = dict([kv.split('=') for kv in unquote(next_results2[1:]).split("&") ])    

    
    search_results2 = twitter_api.search.tweets(**kwargs)
    statuses2 += search_results2['statuses']
    
# Show results
print json.dumps(statuses[0:], indent=1) 
print
print json.dumps(statuses2[0:], indent=1) 
```

### Step #5: Extracting 'coordinates' data for each tweet

Most of the tweets had 'null' coordinates. Therefore, there will be few tweets actually mapped. 
For the 'coordinates' field the longtitude is given first followed by the latitude.***NOTE: The geo field is deprecated thus the coordinate field should be used.***

```python
#Geo Coordinates for q = '#HappyMatturday' 
coordinate_info = [ status['coordinates']
                 for status in statuses]

#Geo Coordinates for q2= '#MakeSomeonesDayIn3Words'
coordinate_info2 = [ status['coordinates'] 
                 for status in statuses2 ]

print " ------GEO INFORMATION FOR #HappyMatturday----"
print json.dumps(coordinate_info[0:], indent=1)
print
print " ------GEO INFORMATION FOR #MakeSomeonesDayIn3Words----"
print json.dumps(coordinate_info2[0:], indent=1)
```

#### Expected Output:

### Step #6: Extracting longitude and latitude values from data

In this step I extracted the actual longitutde and latitude values for my two top trending topics.

```python
#Extracting coordinates from dictionary in object list for '#HappyMatturday' 
lon = []
lat = []
for x in coordinate_info:
    if x != None:
        lon.append(x['coordinates'][0])
        lat.append(x['coordinates'][1])

print "---COORDINATES FOR #HappyMatturday---"
print
for x in range (len(lat)):
    print lon[x], "," ,lat[x]

#Extracting coordinates from dictionary object in list for '#MakeSomeonesDayIn3Words'
lon2 = []
lat2 = []
for x in coordinate_info2:
    if x != None:
        lon2.append(x['coordinates'][0])
        lat2.append(x['coordinates'][1])

print
print "---COORDINATES FOR #MakeSomeonesDayIn3Words---"
print
for x in range (len(lat2)):
    print lon2[x], "," ,lat2[x]
        
```

### Step #7: Plotting coordinates to a map using Google API

Since our search returned mostly null for coordinates I did not have a lot of points to plot. 
In addition, in order to plot multiple markers on a google map you must use the static map API. The embbeded map API only allows for one marker at a time ( I have included, but commented out how to do this for reference ). The actuall google MAPS API can be used to plot multiple markers but it requires javascript and it will no longer be supported soon.

For #HappyMatturday the markers I used were red.
For #MakeSomeonesDayIn3Words the markers I used were green.


***NOTE: To get a static image you must import Image from Ipython.core.display. To use the embbeded map you must import Iframe to allow for the dynamic zooming and display features.***


```python
from IPython.display import IFrame
from IPython.core.display import display
from IPython.core.display import Image

#format our url

#base url
formattedurl = "https://maps.googleapis.com/maps/api/staticmap?"

#add the center of our map (lat,lon), and size required if markers are not present
formattedurl += "size=1000x1000"

#add markers for #HappyMatturday
formattedurl += "&markers=color:red"

for x in range (len(lat)):
    formattedurl += "%7C{},{}".format(lat[x], lon[x])

#add markers for #MakeSomeonesDayIn3Words
formattedurl += "&markers=color:green"

for x in range (len(lat2)):
    formattedurl += "%7C{},{}".format(lat2[x], lon2[x])
    
print formattedurl

#format url for our static map
google_static_map = formattedurl

#display our map
Image(url=google_static_map)

#EMBEDDED GOOGLE MAP EXAMPLE FOR REFERENCE
#google_maps_url = "https://www.google.com/maps/embed/v1/place?key=AIzaSyBpFjRP377IU68WWXHLF8GIVaE8n-ru30U&zoom=2&q=FAU"

#print "---EMBEDDED MAP EXAMPLE---"
#display(IFrame(google_maps_url, '1075px', '500px'))
```
